## Problem 1 

Since we want to analyse historical data, let's work in the time domain. We want a sorted log of all our last pokemon encounters, indexed by time of last encounter.

ðŸ’ª **Task: Load the DataFrame in a variable called `df`, index the DataFrame by `Last Seen` timestamps, then sort the DataFrame in chronological order.**

The `rolling()` method in pandas helps us easily operate on sub periods of your time series(7 days, in our case). In particular, window functions calculate metrics for the **data inside the window**. Then, the result of this calculation forms a **new time series**(`rolling_max_total` in this case), where each data point represents a **summary of several data points** of the original time series. 

<span style='color:Green'> And for documentations, these are some resources you can check out:

## Problem 5

Our pokedex updated its database with new pokemon pictures... but accidentally flipped all of them upside down! Let's fix this:

ðŸ’ªðŸ’ª **Task: Iterate through the pokemon images in the `images` directory, flip them the right side up, then save them in the directory called `images_flipped`.**

Pro-tips:
* to iterate through system files in python, check out [`glob`](https://docs.python.org/3/library/glob.html)
* [`os.path`](https://docs.python.org/3/library/os.path.html) has some great filename utilities


```python
import os                                                                   
import glob
import errno
from PIL import Image
import matplotlib.pyplot as plt

try :
    os.mkdir('./images_flipped')
except OSError as exception:
    if exception.errno != errno.EEXIST:
        raise

for file in list(glob.glob('./images/*.png')):
    image = Image.open(file)
    image_flip = image.transpose(Image.FLIP_TOP_BOTTOM)
    image_flip.save("./images_flipped/" + os.path.basename(file))
```

<span style='color:Green'>ðŸ¤© **functional code & correct method**</span> 

<span style='color:Green'> yup, Pillow to the rescue!
   
<span style='color:Green'> Smart usage of basename to get the name of each image. 
It's cool that you configured an error-handler as well, smart!   

```python
from PIL import Image

def test_image_flip():
    image = Image.open("images_flipped/17.png")
    pixel = image.getpixel((143, 125))
    expected_pixel = (200, 188, 151, 255)
    assert pixel == expected_pixel, f'Image 43.png not flipped. Expected values {expected_pixel} at pixel (143, 125), but got {pixel}'
    print('Success! ðŸŽ‰')


test_image_flip()
```

What is the verb that is most often repeated?

```python
import spacy
nlp = spacy.load("en_core_web_sm")

with open('lyrics.txt', 'r') as f:
    content = f.read()

doc = nlp(content)
```

<span style='color:Green'>ðŸ¤© **functional code & correct method**</span> 

<span style='color:Green'> Great job! `en_core_web_sm` is a small language model that SpaCy provides us with. 
 
<span style='color:Green'>By creating a `doc` object, we can distinguish the parts of speech (POS_) and lemmatize (Lem_) each token!

```python
def test_doc():
    assert "doc" in globals(), "Couldn't find a variable called doc"
    assert len(doc) == 388, f"Expected 388 tokens in doc, but got {len(doc)}"
    print('Success! ðŸŽ‰')
    
test_doc()
```

ðŸ’ªðŸ’ª **Task: Find the most common verb in `doc`**

The code cell below should print out the most common verb:

Pro-tips:
* "VERB" is a [_part of speech_](https://spacy.io/usage/linguistic-features#pos-tagging), which you can directly check on tokens.
* using [_lemmas_](https://spacy.io/usage/linguistic-features#lemmatization) instead of raw token text allows to normalise variations (like verb conjugations)
* [`Counter`](https://pymotw.com/2/collections/counter.html) is an easy way to count repeated elements in collections.

```python
from collections import Counter

verbs = [token.lemma
         for token in doc
         if (not token.is_stop and not token.is_punct and token.pos_ == "VERB")]

verb_freq = Counter(verbs)
print("The most common verb in lyrics.txt is '", [token.text for token in doc if token.lemma == verb_freq.most_common(1)[0][0]][0] , "' and it is used ", verb_freq.most_common(1)[0][1] ," times")
```

<span style='color:Green'>ðŸ¤© **functional code & correct method**</span> 

<span style='color:Green'>smart usage of `.lemma_` and `.pos_` in the list comprehension to store all the verbs in `verb`. I would probably take that last list comprehension out of the print statement, just so that the last line is more legible. You can also check out [python f strings](https://realpython.com/python-f-strings/) for a neat way of printing out variables.

ðŸ’ªðŸ’ª **Bonus Task: Visualise the distribution of verb counts in the lyrics.**

Pro-tip: check out [data to viz](https://www.data-to-viz.com/) to pick the most effective chart type.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 

stopwords = set(STOPWORDS)
verbs_distrib = [token.text
                 for token in doc
                 if (not token.is_stop and not token.is_punct and token.pos_ == "VERB")]
big_string = ' '.join(verbs_distrib)
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(big_string) 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud)
```

But according to Data to viz most optimal way to display this distribution is with bar chart :

```python
from collections import Counter

verbs_freq = Counter(verbs_distrib)
df_verb = pd.DataFrame({'group': verbs_freq.keys(),'values': verbs_freq.values()})
df_verb = df_verb.sort_values(by='values')

plt.hlines(y=range(1,len(df_verb.index)+1), xmin=0, xmax=df_verb['values'])
plt.plot(df_verb['values'], range(1,len(df_verb.index)+1), "o")
plt.yticks(range(1,len(df_verb.index)+1), df_verb['group'])

plt.xlabel('Number of occurence')
plt.ylabel('Verbs')
plt.show()
```
